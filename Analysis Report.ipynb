### Project Synopsis

*Motivation*

Transportation systems in U.S. communities are a critical resource which connect citizens with their jobs, schools, local businesses and community organizations. For most U.S. citizens, transportation systems play a key role towards facilitating nearly all activities within their professional and personal lives. However, engagement with transportation systems carries a level of risk, due to the occurrence of accidents involving personal automobiles.

With this in mind, we were motivated to explore the details and circumstances surrounding such accidents here in Allegheny County. Our team hoped to uncover insights about the nature of vehicular accidents which could potentially inform interventions to improve safety for all who engage with transportation systems in Allegheny County.

*Overview of Data* 

At the outset of the project, our team intended to use a dataset which included data on traffic fatalities in the United States during the years of 2015 and 2016. However, this dataset was only available through the Google BigQuery API, which made it unsuitable for our project. We instead pivoted to another dataset which allowed us to explore our original topic of interest, vehicular accidents. However, this data was more manageable for the scope of this project. 

The new dataset was published by Allegheny County, the City of Pittsburgh and the Western PA Regional Data Center. The dataset provides information on every car accident in Allegheny County which was reported to the police from 2004 - 2017. It contains 190 features about 182,782 car crashes published in one CSV file.

*Methods*

Our project began with data wrangling of the raw data using Python code in a Jupyter notebook ("Project1-DataCleaning-JupyterNotebook.ipynb"). First, we re-set the levels of categorical columns. Next, we split the original CSV file in to four subset data frames, based on themes we discovered in the data: 

- "crash-data.csv": features related to the actual details about the accident
- "driver-data.csv": features related to the driver behavior before or during the accident
- "environment-data.csv": features related to the environmental variables at scene of the accident
- "vehicle-data.csv": features related to the vehicle involved in the accident

After wrangling the data, we proceeded to use Tableau to create data visualizations to discover the story within the data. Our visualizations and findings are outlined below.

#### Key Findings

From our project, we find that: 

- Car accidents follow a seasonal trend; the number of accidents and causes of accidents vary from month to month.
- The number of accidents and causes of accidents also vary by day and time.
- Road conditions and illumination contribute to accidents.
- Aggressive driving and impaired driving contribute to accidents.

More details about the visualizations which informed these findings are listed below.

#### Themes Discovered in the Data

We discovered that the story told by the data about Allegheny County car accidents centers around three key themes:

- *Accident Facts*: How are the basic facts about the accident distributed?
- *Environmental Variables*: How are environmental factors influencing accidents distributed?
- *Driver Behavior*: How are driver behaviors influencing accidents distributed?

The cells below explore each theme in more detail.

### Theme 1: Accident Facts

Theme 1, Accident Facts, considers the basic facts of the accidents within the dataset.

##### Theme 1: Accident Facts: Date and Time 

Our first visualization examines the distribution of accidents across the years within the dataset. We find that the number of accidents is relatively stable from year to year; there is not much of a noticeable trend in variation for the number of accidents from year to year.

Next, we consider the number of accidents as a function of month. This visualization reveals a starker trend; the number of accidents is higher during late fall and winter months (October through February), and lower from spring through late summer months (March through September). Since this trend is so strong, we continue to refer back to it as a point of consideration when considering other aspects of the dataset.

##### Theme 1: Accident Facts: Vehicles Involved 

Our team next considered the type of vehicles involved in the accidents. As this dataset is centered on accidents involving personal automobiles, we were curious how other vehicles might be involved, and if crashes involving other vehicles would follow the same seasonal trend observed in the overall dataset. 

This visualization reveals that for motorcycles and bicycles, the seasonal trend is opposite the seasonal trend observed in the main dataset. Motorcycle and bicycle accidents are highest in the summer months, not the winter months. On the other hand, bus accidents are at their lowest during the summer months (likely due to schools not being in session). 

Trains and trolleys have the lowest number of accidents, while heavy trucks have a consistently high rate of accidents.

##### Theme 1: Accident Facts: Physical Objects

As a final point of our exploration of the accident facts, we were curious about the distribution of objects struck during an accident. This visualization compares the count of object type struck during an accident as a function of month. We observe the dataset's main seasonal trend among most objects, but accidents where a deer was struck are notably higher during the Fall months (September, October and November).

### Theme 2: Environmental Variables

Theme 2 considers the environmental variables at the accident. We consider environmental variables to be features of the accident's physical location which are outside of the driver's control.

###### Theme 2: Environmental Variables: Weather      

The first environmental variable considered was weather. The observation of a strong seasonal trend among the number of crashes led to a hypothesis that weather plays a key role in trend. To learn more, we created a bar chart which considers the number of crashes by month. Each bar is encoded with colors to the type of weather during the accidents.

We had the surprising realization that rain is involved with a much higher number of accidents than snow. After looking at this visualization, this makes sense, as rain occurs during all 12 months of the year, while snow only occurs during certain months.

##### Theme 2: Environmental Variables: Road Conditions                          

After discovering the distribution of accidents among weather types, we were curious what road conditions were associated with accidents, and if these conditions made sense given the trend observed with the weather. We excluded "dry" road conditions, as this was considered to be normal and clear conditions which wouldn't affect the operation of a vehicle. The visualization above shows that wet roads are the leading road condition associated with accidents, being more numerous than all other road conditions combined.

Next, we examined the distribution of road conditions as a function of month. Again, we observe that wet road conditions are the primary type of road conditions associated with an accident, even within the winter months where other conditions like snow and ice are more prominent.

As we considered road conditions, we also were curious if road conditions leading to fatalities were concentrated in any particular location. Discovering any locations with a high occurrence of fatal accidents related to abnormal road conditions could provide an insight that could be useful for officials looking to improve road safety. The visualization above symbolizes the number of fatalities by size of the point, and the type of road condition is symbolized by color.

##### Theme 2: Environmental Variables: School Zones

For our analysis, we considered School Zones to be an environmental variable, as School Zones are special environments where driving conditions deviate from the norm. In particular, School Zones have a much lower speed limit than the surrounding roads. We visualized the number of accidents in a school zone as a function of month, to see how these accidents change with time. It was observed that School Zone accidents follow a seasonal trend, with accidents occurring during the months when schools are in session.

We also made the assumption that School Zones, as a special type of environment, might have a higher prevalance of bicyclists. To understand if collsions with bicyclists in School Zones cooresponded to the Zone's special speed limit, we visulized the number of bicycle-vehicle collsions as a function of speed limit. It appears that the majority of such collisions occured where the driver was following the speed limit, leading us to believe that driver's generally respect School Zones as a special environment by following the lower speed limit.

##### Theme 2: Environmental Variables: Deer Involved

Finally, we consider an environmental variable which is a special consideration in Allegheny County; the prevalence of collisions with deer. We were curious if there were any locations where such collisions occurred frequently. The visualization above shows that accidents involving deer occur frequently on 76, the Pennsylvania Turnpike.

#### Theme 3: Driver Behavior

Theme 3 considers the role of driver behavior in respect to accidents.

##### Theme 3: Driver Behavior: Aggressive Driving

We visualized the percentage of accidents involving aggressive driving and made a surprising and concerning discovery: the percentage of accidents involving aggressive driving in Allegheny County is at least 50% during any given month. October has the highest percentage of accidents involving aggressive driving, with 52.638% of accidents during the month of October involving aggressive driving. 

This finding leads us to believe that an intervention to reduce the prevalence of aggressive driving could be useful towards reducing the number of Accidents in Allegheny County.

##### Theme 3: Driver Behavior: Traffic Violations

Aside from aggressive driving, we were curious if other driver behavior related to accidents. We examined the percentage of other traffic violations, such as running a red light or failure to wear a seat belt, by month. We didn't observe much of a seasonal trend among accidents involving such traffic violations.

##### Theme 3: Driver Behavior: Impaired Driving

One driver behavior with the highest potential for serious consequences is impaired driving. We had a hypothesis that impaired driving may vary strongly by day and time of day, so we created this visualization to explore that hypothesis. We discovered that impaired driving occurs most within the evening hours.

##### Theme 3: Driver Behavior: Distracted Driving

Finally, our team was curious about the variability in the number of accidents due to distracted driving as a function of year. We felt that the number of distracted drivers using cell phones would increase with time over our dataset, as smartphone usage also increased during the same time. Surprisingly, while we found an increase in the number of distracted drivers, we did not see the linear increase in accidents involving cell phone usage which we expected to see.

### Our Recommendations

Based on these visualizations, we make the following recommendations, grouped by theme:

- **Accident Facts**: Interventions to reduce accidents need to be adaptable from season to season, as the number of crashes and details and circumstances surrounding the crash vary from month to month.
- **Environmental Variables**: Weather plays a critical role towards accidents in Allegheny County. While the weather cannot be directly controlled, road illumination and road conditions can be managed to some extent to mitigate the effects of weather on drivers. Interventions to improve illumination and road conditions may reduce the number of accidents in the county.
- **Driver Behavior**: The data shows that aggressive driving is involved in at least half of all accidents in Allegheny County. Public safety campaigns to reduce aggressive driving behavior may reduce the number of accidents in the county. Additionally, impaired driving occurs most frequently at night, so public safety interventions to reduce impaired driving should be increased during evenings.
